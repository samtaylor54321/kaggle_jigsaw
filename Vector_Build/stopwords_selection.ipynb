### Recourring words in this corpus

```python
train = pd.read_csv('../Data/train.csv', usecols=['comment_text'])
test = pd.read_csv('../Data/test.csv', usecols=['comment_text'])
corpus = pd.concat([train, test])
```

```python
corpus['comment_text'] = corpus['comment_text'].str.lower()
corpus.head()
```

```python
potential_additions = [word for comment in corpus['comment_text'].str.split(' ')
                       for word in comment if word not in spacy_stop_words]
```

```python
top_words_in_corpus = pd.Series(potential_additions).value_counts()
top_words_in_corpus.head(20)
```
